```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
main_folder = 'nd280-heavy-neutrino-search-2018_main/'
flux_filename = 'flux.npy'
```

```python
flux = np.load(main_folder+flux_filename)
```

```python
masses = np.linspace(140, 490, 36)
```

```python
plt.plot(masses, flux[:, 10], label='Mode 10 K → µN → µ(e + e − ν(ν̄) µ,τ )')
plt.plot(masses, flux[:, 14], label='Mode 14 K → µN → µ(e + e − ν(ν̄) e )')
plt.xlabel(r'$M_N$ [MeV]')
plt.ylabel('Flux')
plt.legend(frameon=False)
```
