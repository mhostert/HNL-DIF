```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
nd280_limits = np.loadtxt('nd280-heavy-neutrino-search-2018_main/limits_combined/limits_marginalisation.dat')
```

```python
plt.plot(nd280_limits[:, 0], nd280_limits[:, 1], label='$U_e^2$')
plt.plot(nd280_limits[:, 0], nd280_limits[:, 2], label='$U_{\mu}^2$')
plt.plot(nd280_limits[:, 0], nd280_limits[:, 3], label='$U_{\\tau}^2$')
plt.yscale('log')
plt.xlabel(r'$M_N$ [MeV]')
plt.ylabel(r'mixing $U^2$')
plt.legend(frameon=False)
plt.title('Limits obtained after marginalisation')
```

```python
nd280_limits_prof = np.loadtxt('limits_combined/limits_profiling.dat')
```

```python
plt.plot(nd280_limits[:, 0], nd280_limits_prof[:, 1], label='$U_e^2$')
plt.plot(nd280_limits[:, 0], nd280_limits_prof[:, 2], label='$U_{\mu}^2$')
plt.yscale('log')
plt.xlabel(r'$M_N$ [MeV]')
plt.ylabel(r'mixing $U^2$')
plt.legend(frameon=False)
plt.title('Limits obtained after profiling (i.e. setting the other mixings to 0')
```
